### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda3.head()
```

#1. Ventas por categoría



```python
import matplotlib.pyplot as plt
```

```python
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"
```

```python
df = pd.concat([tienda1,tienda2,tienda3,tienda4], ignore_index=True)
df
```

```python
ventas_categoria = df['Categoría del Producto'].value_counts()
ventas_categoria
```

```python
ventas_por_categoria = df.groupby('Categoría del Producto').size()
ventas_por_categoria
plt.figure()
ventas_por_categoria.plot(kind='barh',color='pink',title='Ventas por Categoría')
plt.xlabel('Número de ventas')
plt.ylabel('Categoría')
plt.show()
plt.figure()
ventas_por_categoria.plot(kind='pie',color='pink',title='Ventas por Categoría',autopct='%1.1f%%')
plt.show()
```

# 2. Análisis de facturación

```python

df
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
ingreso_por_tienda = df.groupby('Tienda')['Precio'].sum()
```

```python
plt.figure()
ingreso_por_tienda.plot(kind='pie', title='Ingreso por tienda', autopct='%1.1f%%')
plt.show()
plt.figure()
ingreso_por_tienda.plot(kind='bar', title='Ingreso por tienda')
plt.show()
plt.figure()
ingreso_por_tienda.plot(kind="line", drawstyle="steps")
plt.show()
ingreso_por_tienda
```

```python
df['Utilidad'] = df['Precio'] - df['Costo de envío']
df
```

```python
utilidad_por_tienda = df.groupby('Tienda')['Utilidad'].sum()
print(utilidad_por_tienda)
plt.figure()
utilidad_por_tienda.plot(kind='barh',title='Utilidad por tienda',color='Green')
plt.show()

plt.figure()
utilidad_por_tienda.plot(kind='line',title='Utilidad por tienda',color='Purple')
plt.show()

plt.figure()
utilidad_por_tienda.plot(kind='pie',title='Utilidad por tienda',autopct='%1.1f%%',figsize=(7,7))
plt.show()
```

# 3. Calificación promedio de la tienda


```python
df.head()
```

```python
calificacion_promedio = df.groupby('Tienda')['Calificación'].mean()
calificacion_promedio
```

```python
plt.figure()
calificacion_promedio.plot(
    kind='barh',
    color='green',
    title='Calificación promedio',
    xlim=(0,5),
    ylabel='Tiendas',
    xlabel='Calificación'
)
plt.show()
plt.figure()
calificacion_promedio.plot(
    kind='pie',
    title='Calificación promedio',
    autopct='%1.1f%%'
)
plt.show()
plt.figure()
calificacion_promedio.plot(
    kind='bar',
    color='blue',
    title='Calificación promedio',
    ylim=(0,5),
    xlabel='Tiendas',
    ylabel='Calificación'
)
plt.show()
```

```python
df['Calificación'].plot(kind='kde', title='Densidad de calificaciones')
plt.show()
df['Calificación'].plot(kind='hist', bins=5, title='Distribución de calificaciones')
plt.show()
df.boxplot(column='Calificación', by='Tienda')
plt.title('Distribución de calificaciones por tienda')
plt.suptitle('')
plt.show()
calificacion_promedio.plot(kind='area', title='Calificación promedio')
plt.ylim(0,5)
plt.show()

```
